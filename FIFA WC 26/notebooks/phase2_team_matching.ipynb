```python
import pandas as pd

matches = pd.read_csv("../data/matches_1930_2022.csv")
rankings = pd.read_csv("../data/fifa_ranking_2026-06-08.csv")
```

```python
match_teams = set(matches["home_team"].unique()) | set(matches["away_team"].unique())

ranking_teams = set(rankings["team"].unique())

print("Match Teams:", len(match_teams))
print("Ranking Teams:", len(ranking_teams))
```

```python
missing_in_rankings = sorted(match_teams - ranking_teams)

print("Teams missing in rankings:")
print(missing_in_rankings)
print("\nCount:", len(missing_in_rankings))
```

```python
missing_in_matches = sorted(ranking_teams - match_teams)

print("Teams in rankings but never in World Cup dataset:")
print(missing_in_matches[:50])
print("\nCount:", len(missing_in_matches))
```
