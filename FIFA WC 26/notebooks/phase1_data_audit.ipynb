```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

```python
matches = pd.read_csv("../data/matches_1930_2022.csv")
rankings = pd.read_csv("../data/fifa_ranking_2026-06-08.csv")
world_cup = pd.read_csv("../data/world_cup.csv")
```

```python
print("MATCHES SHAPE")
print(matches.shape)

print("\nRANKINGS SHAPE")
print(rankings.shape)

print("\nWORLD CUP SHAPE")
print(world_cup.shape)
```

```python
print("MATCHES COLUMNS")
print(matches.columns.tolist())

print("\nRANKINGS COLUMNS")
print(rankings.columns.tolist())

print("\nWORLD CUP COLUMNS")
print(world_cup.columns.tolist())
```

```python
print("MATCHES MISSING VALUES")
print(matches.isnull().sum().sort_values(ascending=False).head(20))

print("\nRANKINGS MISSING VALUES")
print(rankings.isnull().sum().sort_values(ascending=False).head(20))

print("\nWORLD CUP MISSING VALUES")
print(world_cup.isnull().sum().sort_values(ascending=False).head(20))
```

```python
print("MATCHES SAMPLE")
display(matches.head())

print("RANKINGS SAMPLE")
display(rankings.head())

print("WORLD CUP SAMPLE")
display(world_cup.head())
```

```python
print("Unique Teams in Rankings")

print(rankings["team"].nunique())
rankings["team"].sort_values().head()
```

```python
print(matches["Year"].min())
print(matches["Year"].max())

print(matches["Round"].value_counts().head(20))
```

```python
summary = pd.DataFrame({
    "Dataset": ["Matches", "Rankings", "World Cup"],
    "Rows": [
        matches.shape[0],
        rankings.shape[0],
        world_cup.shape[0]
    ],
    "Columns": [
        matches.shape[1],
        rankings.shape[1],
        world_cup.shape[1]
    ]
})

summary
```
